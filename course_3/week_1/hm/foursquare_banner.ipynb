```python
meanshift = MeanShift(bandwidth=0.1)
```

```python
meanshift.fit(data[['latitude', 'longitude']])
```

```python
data['claster'] = meanshift.predict(data[['latitude', 'longitude']])
```

```python
a = data.groupby(['claster'])['latitude', 'longitude'].mean()
```

```python
b = data.groupby(['claster'])['id', 'user_id'].count()
```

```python
centers = pd.DataFrame(np.hstack((a, b))).drop(3, axis=1)
```

```python
c15 = centers[centers[2]>15].reset_index()
```

```python
def neight_coordinats(data, coordinats):
    result = []
    for i in range(0, data.shape[0]):
        result.append(data[0][i] - coordinats[0] + data[1][i] - coordinats[1])
    return result
```
